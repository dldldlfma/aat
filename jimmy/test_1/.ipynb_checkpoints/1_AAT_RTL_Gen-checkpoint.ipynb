```python
def con_insert(dic_con, port_mst, port_slv):
    if port_mst.find('.')>0:
        mst_inst, mst_port = port_mst.strip().split('.')
        dic_con[mst_inst][mst_port]['type'] = 'mst'
    else:
        mst_inst, mst_port = '$top', port_mst
        dic_con[mst_inst][mst_port]['type'] = 'top'
    
    if port_slv.find('.')>0:
        slv_inst, slv_port = port_slv.strip().split('.')
        dic_con[slv_inst][slv_port]['type'] = 'slv'
    else:
        slv_inst, slv_port = '$top', port_slv
        dic_con[slv_inst][slv_port]['type'] = 'top'

    # Check Direction
    if dic_con[mst_inst][mst_port]['type']=='$top' or dic_con[slv_inst][slv_port]['type']=='$top':
        if dic_con[mst_inst][mst_port]['dir']== dic_con[slv_inst][slv_port]['dir']=='output':
            pass
        else:
            print ('[LOG] *E, The top connection must have the same mst/slv connection direction.')
            exit()
    else:
        if dic_con[mst_inst][mst_port]['dir']=='output' and dic_con[slv_inst][slv_port]['dir']=='output':
            print ('[LOG] *E, The internal connection must have a differenct mst/slv connection direction.')
            print ('        - mst instance  = ', mst_inst)
            print ('        - mst port      = ', mst_port)
            print ('        - mst direction = ', dic_con[mst_inst][mst_port]['dir'])
            print ('        - slv instance  = ', slv_inst)
            print ('        - slv port      = ', slv_port)
            print ('        - slv direction = ', dic_con[slv_inst][slv_port]['dir'])
            raw_input('Pause ... Press Enter')

    # Check Type
    if dic_con[mst_inst][mst_port]['type']=='mst' and dic_con[slv_inst][slv_port]['type']=='mst':
        print ('[LOG] *E, all mst type')
        exit()
    elif dic_con[mst_inst][mst_port]['type']=='slv' and dic_con[slv_inst][slv_port]['type']=='slv':
        print ('[LOG] *E, all slv type')
        exit()

    print (mst_inst)
    print (mst_port)
    print (slv_inst)
    print (slv_port)
    if dic_con[mst_inst][mst_port]['wire']=='':
        if dic_con[slv_inst][slv_port]['wire']=='':
            tmp_name = mst_inst+'_'+mst_port
        else:
            tmp_name = dic_con[slv_inst][slv_port]['wire']
    else:
        if dic_con[slv_inst][slv_port]['wire']=='':
            tmp_name = dic_con[mst_inst][mst_port]['wire']
        else:
            print ('[LOG] *E, wire name overwrite')
            print ('        - mst instance  = ', mst_inst)
            print ('        - mst port      = ', mst_port)
            print ('        - mst direction = ', dic_con[mst_inst][mst_port]['dir'])
            print ('        - slv instance  = ', slv_inst)
            print ('        - slv port      = ', slv_port)
            print ('        - slv direction = ', dic_con[slv_inst][slv_port]['dir'])
            print ('        - slv direction = ', dic_con[slv_inst][slv_port]['dir'])
            raw_input('Pause ... Press Enter')
            
            
            
    dic_con[mst_inst][mst_port]['wire'] = tmp_name
    dic_con[slv_inst][slv_port]['wire'] = tmp_name
    
    dic_con[mst_inst][mst_port]['ccnt'] = dic_con[mst_inst][mst_port]['ccnt'] + 1
    dic_con[slv_inst][slv_port]['ccnt'] = dic_con[slv_inst][slv_port]['ccnt'] + 1
```

```python
# MAKE Connection Format
#  - Wire Naming 을 위한 Dictionary 생성
#  - Key : Port N
# {'u_mst': {'A': {'ccnt': 0,
#                  'dir': 'output',
#                  'lsb': 0,
#                  'msb': 0,
#                  'role': '',    // mst, slv, top
#                  'wire': '',

TOP_CON = {}

for inst in json_hie['insts']:
    tmp_port = {}
    for port in inst['ports']:
        tmp = {}
        tmp['ccnt'] = 0
        tmp['role'] = ''
        tmp['dir']  = port['dir']
        tmp['msb']  = port['msb']
        tmp['lsb']  = port['lsb']
        tmp['wire'] = ''
        tmp_port[port['name']] = tmp
    TOP_CON[inst['instname']] = tmp_port
    
```

```python
# FROM json_con

# json_con Format
# {'con': [{'mst': {'inst': 'u_mst', 'lsb': 0, 'msb': 0, 'port': 'A'},
#          'slv': {'inst': 'u_slv', 'lsb': 0, 'msb': 0, 'port': 'M'},
#          'type': 'p2p'},

# TOP_CON = {'u_mst': {'A': {'ccnt': 0,
#                             'dir': 'output',
#                             'lsb': 0,
#                             'msb': 0,
#                            'role': '',    # mst, slv, top
#                            'wire': ''},


# role 설정 및 Check (mst, slv, top)
# ccnt (connection count) 설정 : 1 이상이면 wire 선언

for con in json_con['con']:
    mst_inst = con['mst']['inst']
    mst_port = con['mst']['port']
    mst_lsb  = con['mst']['lsb']
    mst_msb  = con['mst']['msb']
    mst_dir  = TOP_CON[mst_inst][mst_port]['dir']
    slv_inst = con['slv']['inst']
    slv_port = con['slv']['port']
    slv_lsb  = con['slv']['lsb']
    slv_msb  = con['slv']['msb']
    slv_dir  = TOP_CON[slv_inst][slv_port]['dir']

    # Role Config
    if TOP_CON[mst_inst][mst_port]['role']=='':
        if mst_inst=='$top': mst_role = 'top'
        else               : mst_role = 'mst'
        TOP_CON[mst_inst][mst_port]['role'] = mst_role
    elif TOP_CON[mst_inst][mst_port]['role']== 'slv':
        print ('[LOG] *E, Port role has been changed')
        print ('       - mst -> slv')
        exit()

    if TOP_CON[slv_inst][slv_port]['role']=='':
        if slv_inst=='$top': slv_role = 'top'
        else               : slv_role = 'slv'
        TOP_CON[slv_inst][slv_port]['role'] = slv_role
    elif TOP_CON[slv_inst][slv_port]['role']=='mst':
        print ('[LOG] *E, Port role has been changed')
        print ('       - slv -> mst')
        exit()

    # Check Direction
    if (mst_inst=='$top') ^ (slv_inst=='$top'):
        if mst_dir==slv_dir:
            pass
        else:
            print ('[LOG] *E, The top connection must have the same mst/slv connection direction.')
            exit()
    else:
        if mst_dir=='output' and slv_dir=='output':
            print ('[LOG] *E, The internal connection must have a differenct mst/slv connection direction.')
            print ('        - mst instance  = ', mst_inst)
            print ('        - mst port      = ', mst_port)
            print ('        - mst direction = ', mst_dir )
            print ('        - slv instance  = ', slv_inst)
            print ('        - slv port      = ', slv_port)
            print ('        - slv direction = ', slv_dir )
            raw_input('Pause ... Press Enter')

    # Write Connection Counter
    TOP_CON[mst_inst][mst_port]['ccnt'] = TOP_CON[mst_inst][mst_port]['ccnt'] + 1
    TOP_CON[slv_inst][slv_port]['ccnt'] = TOP_CON[slv_inst][slv_port]['ccnt'] + 1
    
pprint (TOP_CON)
```

```python
# TOP_CON = {'u_mst': {'A': {'ccnt': 0,
#                             'dir': 'output',
#                             'lsb': 0,
#                             'msb': 0,
#                            'role': '',    # mst, slv, top
#                            'wire': ''}

# TOP_WIRE = [ ['wire_name',lsb, msb], , , ]
TOP_WIRE_MST = []
TOP_WIRE_SLV = []

# TOP_ASSIGN = [ ['assign_name',lsb, msb, var_name, var_lsb, var_msb], , , ]
TOP_ASSIGN = []


# wire 설정
# flag 설정



for con in json_con['con']:
    mst_inst = con['mst']['inst']
    mst_port = con['mst']['port']
    mst_lsb  = con['mst']['lsb']
    mst_msb  = con['mst']['msb']
#    mst_dir  = TOP_CON[mst_inst][mst_port]['dir']
    mst_wire = TOP_CON[mst_inst][mst_port]['wire']
    mst_dir  = TOP_CON[mst_inst][mst_port]['dir']
    slv_inst = con['slv']['inst']
    slv_port = con['slv']['port']
    slv_lsb  = con['slv']['lsb']
    slv_msb  = con['slv']['msb']
#    slv_dir  = TOP_CON[slv_inst][slv_port]['dir']
    slv_wire = TOP_CON[slv_inst][slv_port]['wire']
    slv_dir  = TOP_CON[slv_inst][slv_port]['dir']

    mst_wire_name = mst_inst+'_'+mst_port
    slv_wire_name = slv_inst+'_'+slv_port
    
    if mst_wire=='':
        TOP_WIRE_MST.append ([mst_wire_name, mst_lsb, mst_msb])
        TOP_CON[mst_inst][mst_port]['wire'] = mst_wire_name

    if slv_wire=='':
        if TOP_CON[slv_inst][slv_port]['ccnt']>1:
            TOP_WIRE_SLV.append ([slv_wire_name, slv_lsb, slv_msb])
            TOP_CON[slv_inst][slv_port]['wire'] = slv_wire_name
            if mst_dir=='output':
                TOP_ASSIGN.append ([slv_wire_name, slv_lsb, slv_msb, mst_wire_name, mst_lsb, mst_msb])
            else:
                TOP_ASSIGN.append ([mst_wire_name, mst_lsb, mst_msb, slv_wire_name, slv_lsb, slv_msb])
        else:
            TOP_CON[slv_inst][slv_port]['wire'] = mst_wire_name
    
    


pprint (TOP_CON)
```
